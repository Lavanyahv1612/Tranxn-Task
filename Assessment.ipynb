```python
df_read = pd.read_excel("cleaned_file.xlsx")

col_name = "REPAIR_DLR_POSTAL_CD"

def convert_ASCII(s):
    if pd.isna(s):
        return ""
    s = str(s).replace(" ", "")
    result = ""
    for c in s:
        if c.isalpha():
            result += str(ord(c))
        elif c.isdigit():
            result += c
    return result

df_read["REPAIR_DLR_POSTAL_CD"] = df_read[col_name].apply(convert_ASCII)

print("Conversion complete! Saved/n")
```
